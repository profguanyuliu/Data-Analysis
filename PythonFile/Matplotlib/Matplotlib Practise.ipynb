# scatter practise case
The temps of November in Beijing

```python
from matplotlib import pyplot as plt
from matplotlib import font_manager

#make the data
y = [11,17,16,11,12,11,12,6,6,7,8,9,12,15,14]
x = range(1,16)

#size the pic
plt.figure(figsize=(10,4),dpi=80)

plt.scatter(x,y,label="November")

#set x axis
_x = list(x)
_xtick_labels = ["Nov{}".format(i) for i in x]
plt.xticks(_x,_xtick_labels,rotation=45)

#set y axis
_y = list(y)
_ytick_labels = ["{}'C".format(i) for i in y]
plt.yticks(_y,_ytick_labels)

#set the grid
plt.grid(alpha=0.4)

#add the legend
plt.legend(loc="upper right")

#describe the pic
plt.xlabel("Time")
plt.ylabel("Degree")
plt.title("Temps in November")

#save the pic
plt.savefig("./scatter.png")

#draw the pic
plt.show()
```

# bar practise case

```python
import matplotlib.pyplot as plt

#make the data
x = ["Math","ComS","Infosys","Fintech"]
y = [50,35,65,40]

#size the pic
plt.figure(figsize=(10,4),dpi=80)

plt.bar(x,y,label="Studentnum in SEM",width=0.3,color="pink")

#set x axis
plt.xticks(color="black")

#set y axis
plt.yticks(color="black")

#set the grid
#plt.grid(alpha=0.4)

#add the legend
plt.legend(loc="upper left")

#describe the pic
plt.xlabel("Departments in SEM")
plt.ylabel("Student Number")
plt.title("Student Number of SEM")

#save the pic
plt.savefig("./bar.png")

plt.show()
```

# plot practise case
